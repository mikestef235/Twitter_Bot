```python
import pandas as pd
from collections import OrderedDict
```

```python
f1 = open('final_questions.txt', 'r').readlines()
f2 = open('answers.txt', 'r').readlines()
```

```python
#Setting the questions and answers generated by the bot
questions = [line.strip('\n') for line in f1]
answers = [line.strip('\n') for line in f2]

#Building the Q+A df
df = pd.DataFrame(data=OrderedDict({'Question': questions, 'Answer': answers}))
df = df[~(df.Answer.str.contains('unk') | df.Answer.str.contains('wiki'))]
```

```python
df.to_csv('QandADf.csv', encoding='utf8')
```
